```python
import pandas as pd
from datetime import datetime
import requests


hour = datetime.utcnow().hour

token =

header = {
    'X-Mapy-Api-Key': token
}

obce = pd.read_csv('in/tables/Obce_kancelare_oprava.csv')
obce = obce.sample(frac=1)
trasy = []

for _, row in obce.iterrows():
    try:
        spicka = "ranni"
        data = {
             "start": [row["souradnice_y"], row["souradnice_x"]],        
             "end": [row["Y_KANCELAR"], row["X_KANCELAR"]],
             "routeType": 'car_fast'
         }
            
        cas_stazeni = datetime.utcnow().strftime('%Y-%m-%d %H:%M:%S')
        response = requests.get('https://api.mapy.com/v1/routing/route', headers=header, params=data, timeout=30)
        response = response.json()
        trasy.append({
          			"CAS_STAZENI": cas_stazeni,
          			"SPICKA": spicka,
                "OBEC_KOD": row["OBEC_KOD"],
          			"OBEC_JMENO": row["OBEC_NAZEV"],
                "KANCELAR_JMENO": row["JMENO"],
                "DELKA": response["length"],
                "DOBA": response["duration"],
          			"SOURADNICE": response["geometry"]["geometry"]["coordinates"]
            })
    except Exception:
        pass
trasy = pd.DataFrame(trasy)

trasy['ID'] = range(1, len(trasy) + 1)
trasy.to_csv("out/tables/trasy_kancelare_rozsirene_druhy_tyden_bez_dopravy.csv", index = False)
```
